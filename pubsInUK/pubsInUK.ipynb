```python
import folium
from folium.plugins import MarkerCluster
import geopandas as gpd
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('C:\\Users\\actua\\Desktop\\DataSets\\pubs\\open_pubs.csv')
df
```

```python
df.isnull().sum()
df = df.drop(columns= ['local_authority'])
df
```

```python
latitude = 52.3555
longitude = -1.1743
myMap = folium.Map(location = [latitude,longitude], zoom_start = 6.5)

myMap
```

```python
#formatting for the mapmarker popup
popup = folium.Popup(min_width = 500)
```

```python
for x in range(df.shape[0]):
    latitudeForPubs = float(df.iloc[x]['latitude'])
    longitudeForPubs = float(df.iloc[x]['longitude'])
    popUp = popUp = ("FSA ID: "+ str(df.iloc[x]['fas_id']) + '\n' + str(df.iloc[x]['name']) + '\n' + str(df.iloc[x]['address']) + '\n' + str(df.iloc[x]['postcode']))
    print(x,latitudeForPubs,longitudeForPubs)
    folium.Marker([latitudeForPubs,longitudeForPubs], popup = popUp).add_to(myMap)
```
