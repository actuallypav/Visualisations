```python
import folium
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:\\Users\\actua\\Desktop\\DataSets\\breweries\\breweries_us.csv')
df = df.rename(columns={"state_breweries":"count"})
df['count'] = 1
```

```python
df
```

```python
amountOfBreweries = df.groupby('state',as_index=False,sort=False).sum()
amountOfBreweries
```

```python
typeOfBrewery = pd.pivot_table(df, values = ['count'], index = ('state','type'), aggfunc=np.sum)
typeOfBrewery = typeOfBrewery.rename(columns={"count":"amount"})
typeOfBrewery
```

```python
geoUS = gpd.read_file("C:\\Users\\actua\\Desktop\\geos\\us-states.geojson")
geoUS = geoUS.drop(geoUS.index[[24]])
geoUS = geoUS.rename(columns={"name":"state"})
geoUS
```

```python
geoShapesUS = geoUS[['state','geometry']]
geoShapesUS
```

```python
geoAndAmount = geoShapesUS.merge(amountOfBreweries,on='state')
geoAndAmount
```

```python
xMap= geoUS.longitude.astype(float).mean()
yMap= geoUS.latitude.astype(float).mean()
print(xMap, yMap)
```

```python
theMap = folium.Map(location=[yMap,xMap],zoom_start=4,tiles=None)
folium.TileLayer('CartoDB positron',name='Light Map',control=False).add_to(theMap)
```

```python
#creating choropleth map

m = folium.Map([yMap, xMap], zoom_start=3)

folium.Choropleth(
    geo_data=geoAndAmount,
    data=geoAndAmount,
    columns=['state', 'count'],
    fill_color='YlOrRd',
    fill_opacity=0.2, 
    line_opacity=0.8,
    legend_name='Total number of breweries per state',
    key_on='feature.properties.state',
).add_to(m)

m
```

```python
style_function = lambda x: {'fillColor': '#ffffff', 
                            'color':'#000000', 
                            'fillOpacity': 0.1, 
                            'weight': 0.1}

state = folium.features.GeoJson(
    geoAndAmount,
    style_function=style_function, 
    control=False,
    tooltip=folium.features.GeoJsonTooltip(
        fields=['state',
                'count'
               ],
        aliases=['State: ',
                 'Total number of breweries: '
                 ],
        style=("background-color: white; color: #333333; font-family: arial; font-size: 12px; padding: 10px;") 
    )
)
m.add_child(state)
m.keep_in_front(state)
folium.LayerControl().add_to(m)
```
